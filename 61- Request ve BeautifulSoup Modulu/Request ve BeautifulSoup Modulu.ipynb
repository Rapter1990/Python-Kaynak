```python
#request import ettik
import requests

#beautifulsoup import ettik
from bs4 import BeautifulSoup

#BeautifulSoup sayseinde tagları alabiliriz.

url =  "https://yellowpages.com.tr/ara?q=Ankara" # Site linkimiz

response = requests.get(url) # Web sayfamızı çekiyoruz.

#print(response) # 200 web sayfasından bilgi çekme ve tüm bilgiler response içinde

html_icerik = response.content # # Web sayfamızın içeriğini alıyoruz.(html kaynak kodu)

soup = BeautifulSoup(html_icerik,"html.parser") # Web sayfamızı parçalamak için BeautifulSoup objesine ve parser ile parçalayıp atıyoruz.

#print(soup.prettify()) # Güzel görüntü için prettify() fonksiyonunu kullanıyoruz. Tüm html kaynak kodu gösterdi

#print(soup.find_all("a")) # tüm <a> tag larını liste hailnde gösterir.

# tüm <a> taglarını gösterme
# for i in soup.find_all("a"):
#     print(i)
#     print("\n********************************************************")


# tüm <a> tagların sadece href olan kısımlardaki değerleri alma
# for i in soup.find_all("a"):
#     print(i.get("href"))
#     print("\n****************************************")



#tüm <a> aglarındaki yazılar için
# for i in soup.find_all("a"):
#     print(i.text)
#     print("\n****************************************")



# divdeki class yp-boi-box-2 olanları gösterecek
#print(soup.find_all("div",{"class":"yp-poi-box-2"})) -> liste olarak döndürdü

# Bu kullanımın anlamı div etiketlerinden class'ı yp-poi-box-2 yi al anlamına geliyor.
for i in soup.find_all("div",{"class":"yp-poi-box-2"}):
    print(i)

```
